# Parse Json File

```python
import json
```

```python
# parse json scrapped from twitter
input_file = open('mydata.json','r')
input_raw = input_file.read()
input_json = json.loads(input_raw)
```

```python
# number of tweets we got
len(input_json)
```

# Extract Sentence and Tokenization

```python
sentences = []
words = []

for i in input_json:
    sentences.append(i['tweet_text'])
    # only keep English words
    for j in i['tweet_text'].split():
        if(j.isascii()):
            words.append(j)
```

```python
# number of words we got
len(words)
```

```python
words
```

```python
sentences
```

# Remove punctuation

```python
import string
```

```python
#Show the punctuation
string.punctuation
```

## There are special cases that we don't want to delete the punctuation
1. Mr. Fang
2. Y. W. Fang

```python
en_titles = ['Mr.', 'Mrs.', 'Dr.', 'Prof.','Hon.']

for i in range(len(words)):
    if words[i] in en_titles:
        continue
    elif len(words[i])==2 and words[i][0].isupper():
        continue
    else:
        for j in string.punctuation:
            words[i] = words[i].replace(j, '')
```

```python
print(words)
```

```python
# remove blank string from the list
words = [i for i in words if i != ""]
```

```python
print(words)
```

# Each character convert to lower case

```python
words = [i.lower() for i in words]
```

```python
print(words)
```

# Delete Stop words

```python
# define stop words
f = open('stop_words.txt')
stop_words = f.read().split('\n')
print(stop_words)
```

```python
# now we create a new word set without the stop words
words = [words[i] for i in range(len(words)) if words[i] not in stop_words]

len(words)
```

```python
print(words)
```

# Stemming

```python
from nltk.stem import PorterStemmer 
   
ps = PorterStemmer() 

for i in range(len(words)):
    words[i] = ps.stem(words[i])
```
